```python
raw = []        # [] 이건 list를 만드는 것
seq = ""
name = ""       # 어차피 for loop 돌리면서 .append에 { key값과 value값 형태로 } 저장후에 데이터프레임 만들것임
```

name = line[1:] 은 글자 첫번째 ">"를 저장하지 않기 위함으로,
만약 >를 포함해 저장하고자 한다면
name = line
혹은 
name = line[0:]

seq = "" 이 없으면, 다음 seq += line을 읽을 때, 이전 line의 값이 누적으로 들어가게 된다.

또한, 맨 마지막에 raw.append로 한번 더 추가하는 이유는, 마지막 줄이 누락되지 않게 하기 위함이다.

```python
with open('C:/Python27/rosalind_p/rosalind_gc.txt',"r") as gene:
    for line in gene:
        line = line.strip()     # line으로 읽을 때, 불필요한 /n이나 /t 존재 제거 (특정 문자를 제거하고 싶으면 .strip(">")도 가능)
        if line.startswith(">"):
            if name:
                raw.append({"name":name,"seq":seq})     # .append로 지정해주지 않으면, 값이 최신 값으로 갱신되어 저장됨 즉, 마지막 줄만 보여줌
            name = line[1:]
            seq = ""
        else:
            seq += line

if name:
    raw.append({"name":name,"seq":seq})     # 위에 말했던 것처럼, 처음부터 읽어서 마지막줄만 넣어줌
```

```python
raw = pd.DataFrame(raw)
raw
```

```python
raw["count"] = raw["seq"].apply(lambda x: x.count("G") + x.count("C"))
raw
```

```python
raw["GCcontents"] = raw["count"]/raw["seq"].apply(lambda x: len(x))*100
raw
```

```python
target = raw[raw["GCcontents"] == raw["GCcontents"].max()]
target
```

```python
result = f"{target['name'].values[0]}\n{target['GCcontents'].values[0]}"
print(result)
```
